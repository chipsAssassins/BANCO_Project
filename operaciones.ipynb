```python

import ezdxf
import numpy as np

f = lambda x: -x**4 + 5*x**2 + x + 9
g = lambda x: 0.4*x**3 - 5*x
h = lambda x: f(x) - g(x)

def bisect(a, b, tol=1e-9):
    fa, fb = h(a), h(b)
    if fa * fb > 0: return None
    for _ in range(100):
        mid = (a + b) / 2
        fm = h(mid)
        if abs(fm) < tol: break
        if fa * fm < 0: b, fb = mid, fm
        else: a, fa = mid, fm
    return mid

# Find intersection x-values
xMin, xMax = -4, 4
steps = 10000
dx = (xMax - xMin) / steps
roots = []
prev = h(xMin)
for i in range(1, steps+1):
    x = xMin + i * dx
    cur = h(x)
    if prev * cur < 0:
        r = bisect(x - dx, x)
        if r is not None and (not roots or abs(r - roots[-1]) > 0.05):
            roots.append(r)
    prev = cur

print(f"Intersecciones en x: {[round(r,5) for r in roots]}")

# For each enclosed region between consecutive roots, build closed polyline
doc = ezdxf.new('R2010')
msp = doc.modelspace()

scale = 10  # 1 unit = 10mm

for i in range(len(roots)-1):
    x0, x1 = roots[i], roots[i+1]
    # Check if f > g in this interval (region above g, below f)
    xmid = (x0 + x1) / 2
    if h(xmid) < 0:
        x0, x1 = x1, x0  # swap if needed — just skip reversed intervals
        continue
    
    n = 500
    xs = np.linspace(x0, x1, n)
    
    # Top curve (f), then bottom curve (g) reversed
    top = [(x * scale, f(x) * scale) for x in xs]
    bot = [(x * scale, g(x) * scale) for x in reversed(xs)]
    
    points = top + bot + [top[0]]  # close the loop
    msp.add_lwpolyline(points, close=True)
    
path = r"C:\Users\Angel Alberto\Downloads\curvas_region.dxf"
doc.saveas(path)
print(f"DXF guardado en: {path}")
```
